# Обработка экспериментальных данных

```python
def linear_func(x, a, b):
    return a * x + b

data = pd.read_excel("data.xlsx")


T = data["T"][:-1]
U = data["U"][:-1]
I = data["I"][:-1]

plt.figure(figsize=(5, 5))
plt.scatter(U, I, label='Экспериментальные данные', color='b')
popt, pcov = curve_fit(linear_func, U, I)
plt.plot(U, linear_func(U, *popt), 'r-', label='Аппроксимация: a=%.3f, b=%.3f' % tuple(popt))
plt.xlabel("U")
plt.ylabel("I")
plt.title("Аппроксимация ")
plt.legend()
plt.show()

true_w1 = 2.8
true_w0 = 1.5
xs = np.arange(0, 10, 0.5)
noise = np.random.normal(0, 0.45, size=xs.shape)
ys = true_w1 * xs + true_w0 + noise

plt.figure(figsize=(5, 5))
plt.scatter(xs, ys, label='с шумом', color='g')
popt, pcov = curve_fit(linear_func, xs, ys)
plt.plot(xs, linear_func(xs, *popt), 'r-', label='Аппроксимация: a=%.3f, b=%.3f, \n true_w1 = 2.8, true_w0 = 1.5' % tuple(popt), )
plt.xlabel("x")
plt.ylabel("y")
plt.title("Аппроксимация с шумом")
plt.legend()
plt.show()
```

# Правдоподобие для гауссовой вероятностной модели

```python
mu = 5
sigma = 2
N = 100
xs = np.random.normal(mu, sigma, size=N)

mu_exp = xs.mean()
sigma_exp = (xs.var()) ** (1/2)
print("Параметры выборки: mu = ", mu, "sigma = ", sigma)
print("Аппроксимация: mu = ", mu_exp, "sigma = ", sigma_exp)
```
