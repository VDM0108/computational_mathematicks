# 2. Американские горки

```python
import numpy as np

def f6(x):
  return (x-5)*np.exp(x)

def f2(x):
  return x**3/3+x**2

eps = 0.01
a = -1
b = 1
while (b-a)>eps:
  c = (b-a)/2
  f_left = f6(c-eps)
  f_right = f6(c+eps)
  if f_left < f_right:
    b = c
  else:
    a = c
print('Минимум f6 (дихотомия): ',(a+b)/2)

a = -1
b = 0.5
N = 100
fib = [1,1]
for i in range(2,N+1):
  fib.append(fib[i-1]+fib[i-2])
x1 = a+(b-a)*fib[N-2]/fib[N]
x2 = a+(b-a)*fib[N-1]/fib[N]
f_1 = f2(x1)
f_2 = f2(x2)
for i in range(N,0,-1):
  if (x2-x1)<1e-8:
    break
  if f_1 > f_2:
    a = x1
    x1 = x2
    x2 = b-(x1-a)
    f_1 = f_2
    f_2 = f2(x2)
  else:
    b = x2
    x2 = x1
    x1 = a+(b-x2)
    f_2 = f_1
    f_1 = f2(x1)
print('Минимум f2 (Фибоначчи): ',(x1+x2)/2)
plt.plot(np.linspace(-10, 10, 100), f6(np.linspace(-10, 10, 100)))
plt.title('f6')
plt.show()
plt.plot(np.linspace(-10, 10, 100), f2(np.linspace(-10, 10, 100)))
plt.title('f2')
plt.show()
```

# 3. Функция Растригина

```python
from scipy.optimize import minimize, Bounds
from math import pi

def func(x):
  return 20 + x[0]**2 + x[1]**2 - 10*np.cos(2*pi*x[0]) - 10*np.cos(2*pi*x[1])

def func1(x):
  return (-1)*(20 + x[0]**2 + x[1]**2 - 10*np.cos(2*pi*x[0]) - 10*np.cos(2*pi*x[1]))

mins_x = []
mins_y = []
maxs_x = []
maxs_y = []
for i in range(-5,6,1):
  for j in range(-5,6,1):
    bounds = Bounds ([i-0.5, j-0.5], [i+0.5, j+0.5])
    mins = minimize(func,(i,j),bounds=bounds)['x']
    mins_x.append(mins[0])
    mins_y.append(mins[1])
for i in range(-5,5,1):
  for j in range(-5,5,1):
    bounds = Bounds ([i, j], [i+1, j+1])
    maxs = minimize(func1,(i+0.5,j+0.5),bounds=bounds)['x']
    maxs_x.append(maxs[0])
    maxs_y.append(maxs[1])

plt.scatter(mins_x,mins_y)
plt.scatter(maxs_x,maxs_y,color='r')
```

# 5. Овальный

Если функция овражная, то, когда алгоритм ищет минимум вдоль градиента, он будет проходить малое расстояние. Чтобы дойти до минимума (как на рисунке), нужно пройти вдоль внутреннего овала (т.к. так работает метод), а это займет много итераций, потому что вектора в данном методе получаются перпендикулярными.

Если аппроксимировать функцию около минимума квадратичной формой, то наименьшее сингулярное число будет значительно меньше других, так как по условию линии сильно вытянуты в одном направлении. Это направление и будет соответствовать наименьшему сингулярному числу.

При взятии обратной матрицы из наименьшего сингулярного числа получится наибольшее, поэтому наибольшее сингулярное число $||A^{-1}||$ будет существенно больше других сингулярных чисел. При этом, так как $||A||=\sqrt{\max \lambda}$, то получится, что число обусловленности $K = ||A|| \cdot ||A^{-1}|| \gg 1$.
