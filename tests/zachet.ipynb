```python
import numpy as np

# Начальные условия
x, y, z = 0, 0, 0  
tolerance = 0.01   
max_iterations = 100  

def simple_iterations(x, y, z):
    x_new = (3 - 5*y - 5*z) / 24
    y_new = (5 - x - 2*z) / 21
    z_new = (9 - 4*x - 6*y) / 19
    return x_new, y_new, z_new

# Итерационный процесс
iteration = 0
errors = []

while iteration < max_iterations:
    x_new, y_new, z_new = simple_iterations(x, y, z)
    
    # условие остановки
    error_x = abs(x_new - x)
    error_y = abs(y_new - y)
    error_z = abs(z_new - z)
    errors.append((error_x, error_y, error_z))
    
    if error_x <= tolerance and error_y <= tolerance and error_z <= tolerance:
        break
    
    x, y, z = x_new, y_new, z_new
    iteration += 1

# Вывод результатов
print(f"Решение достигнуто за {iteration} итераций:")
print(f"x ≈ {x:.4f}, y ≈ {y:.4f}, z ≈ {z:.4f}")
print(f"Погрешности на последней итерации: Δx ≈ {error_x:.4f}, Δy ≈ {error_y:.4f}, Δz ≈ {error_z:.4f}")
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
 
# Определение функции для метода простых итераций для каждого варианта 
def method_a(x, tol=1e-6, max_iter=100): 
    """Метод a: x_{n+1} = x_n - (x_n^3 - 4x_n - 4)/15""" 
    iter_count = 0 
    while iter_count < max_iter: 
        x_next = x - (x**3 - 4*x - 4) / 15 
        if abs(x_next - x) < tol: 
            return x_next, iter_count 
        x = x_next 
        iter_count += 1 
    return x, iter_count 
 
def method_b(x, tol=1e-6, max_iter=100): 
    """Метод b: x_{n+1} = 2 * sqrt(1 + 1/x_n)""" 
    iter_count = 0 
    while iter_count < max_iter: 
        x_next = 2 * np.sqrt(1 + 1 / x) 
        if abs(x_next - x) < tol: 
            return x_next, iter_count 
        x = x_next 
        iter_count += 1 
    return x, iter_count 
 
def method_c(x, tol=1e-6, max_iter=100): 
    """Метод c: x_{n+1} = (x_n^3)/4 - 1""" 
    iter_count = 0 
    while iter_count < max_iter: 
        x_next = (x**3) / 4 - 1 
        if abs(x_next - x) < tol: 
            return x_next, iter_count 
        x = x_next 
        iter_count += 1 
    return x, iter_count 
 
# Начальная точка 
x0 = 2.0 
 
# Решение для каждого метода 
result_a, iter_a = method_a(x0) 
result_b, iter_b = method_b(x0) 
# result_c, iter_c = method_c(x0) 
 
# Вывод результатов 
print(f"Метод A: корень = {result_a}, итерации = {iter_a}") 
print(f"Метод B: корень = {result_b}, итерации = {iter_b}") 
# print(f"Метод C: корень = {result_c}, итерации = {iter_c}") 
 
# График сходимости 
iterations = np.arange(max(iter_a, iter_b)) 
 
# Метод A 
errors_a = [abs(method_a(x0 + i*0.1)[0] - result_a) for i in iterations] 
# Метод B 
errors_b = [abs(method_b(x0 + i*0.1)[0] - result_b) for i in iterations] 
# Метод C 
# errors_c = [abs(method_c(x0 + i*0.1)[0] - result_c) for i in iterations] 
 
plt.figure(figsize=(10, 6)) 
plt.plot(iterations, errors_a, label="Метод A", marker='o') 
plt.plot(iterations, errors_b, label="Метод B", marker='x') 
# plt.plot(iterations, errors_c, label="Метод C", marker='s') 
plt.yscale('log') 
plt.xlabel('Итерации') 
plt.ylabel('Ошибки') 
plt.title('Сходимость методов простых итераций') 
plt.legend() 
plt.grid(True) 
plt.show()
```
