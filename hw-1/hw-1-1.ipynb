```python
import numpy as np
import matplotlib.pyplot as plt
```

# Numpy

```python
m1 = np.array([[1, 2], [3, 4]])
m2 = np.array([[5], [6]])

combined_matrix = np.hstack((m1, m2))

combined_matrix
```

```python
M = np.array([[1, 2], [3, 4]])
a = np.array([0, 1, 0])
b = np.array([1, 0, 0])

result_array = M[a, b]

result_array
```

```python
m = np.array([[1, 2], [3, 4]])

trace = np.trace(m)

nonzero_diag_elements = np.count_nonzero(np.diagonal(m))

print(trace)
print(nonzero_diag_elements)
```

```python
m1 = np.array([[1, -1, 3], [-2, -2, 2]])
m2 = np.array([[1, 2, 3], [4, 5, 6]])

m2[m1 < 0] = 0

m2
```

```python
m1 = np.array([[1, 2], [3, 4], [5, 6]])
m2 = np.array([[7, 8], [9, 10], [11, 12]])

matrix_products = np.dot(m1.T, m2)

matrix_products
```

# Matplotlib

```python
fig, axes = plt.subplots(2, 2)

x = np.linspace(0, 10, 100)
axes[0, 0].plot(x, np.sin(x))
axes[0, 1].plot(x, np.cos(x))
axes[1, 0].plot(x, x**2)
axes[1, 1].plot(x, x)

plt.show()
```

```python
x = np.linspace(0, 10, 10)
y = np.cos(x)
error_y = np.random.rand(10) * 0.2
error_x = np.random.rand(10) * 0.5

plt.errorbar(x, y, yerr=error_y, xerr=error_x, fmt='o')
plt.plot(np.linspace(0, 10, 100), np.cos(np.linspace(0, 10, 100)), linestyle='dashed')
plt.title('Error Bar Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.show()
```

```python
x_log = np.logspace(0.1, 2, 100) 
y_log = np.logspace(5, 10, 100) 
plt.loglog(x_log, y_log)
plt.title('Log-Log Plot')
plt.xlabel('X(log scale)')
plt.ylabel('Y(log scale)')

plt.show()
```
